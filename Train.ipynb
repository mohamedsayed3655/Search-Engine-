```python
import os
from searc_engine import EgyptianFiguresSearchEngine

CORPUS_PATH='corpus (1).json'
MODEL_DIR="models"

os.makedirs(MODEL_DIR, exist_ok=True)
def main():
    print("=" * 50)
    print("  Egyptian Figures Search Engine — Trainer")
    print("=" * 50)

    engine = EgyptianFiguresSearchEngine()
    engine.load_corpus(CORPUS_PATH)
    engine.fit()
    engine.save(MODEL_DIR)

    s = engine.stats()
    print(f"\n  Documents : {s['num_documents']}")
    print(f"  Vocab     : {s['vocab_size']} terms")
    print(f"  Avg len   : {s['avg_doc_length']} tokens")
    print(f"  Fields    : {', '.join(sorted(s['fields']))}")
    print(f"\n  Model saved to: {MODEL_DIR}/search_engine.pkl")
    print("\n  Run the app with:")
    print("    streamlit run interface/app.py")
    print("=" * 50)

if __name__ == "__main__":
    main()
```
